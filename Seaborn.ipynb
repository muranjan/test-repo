```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Available options are "darkgrid" / "white" / "ticks"/ "whitegrid" -- Default is darkgrid
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# "relplot" is Figure-level interface for drawing relational plots onto a FacetGrid.
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
# We can use a third variable in "hue" to differentiate the color of x and y 
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker")
```

```python
# This is called Categorical Pallets
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker", style="time")
```

```python
# This is called Qualitative Pallets
sns.relplot(x="total_bill",y="tip",data=tips,hue="size")
```

```python
# we can use "palette" param to use Colors for the different levels of the ``hue`` variable
sns.relplot(x="total_bill",y="tip",data=tips,hue="size", palette='ch:r=-0.2, l=0.95')
```

```python
# we can use "size" param to increase the size of the circles
sns.relplot(x="total_bill",y="tip",data=tips,size="size")
```

```python
# we can use "sizes" param to normalize the size of "size" param
sns.relplot(x="total_bill",y="tip",data=tips,size="size",sizes=(50,200))
```

```python
# we can use "sizes" param to normalize the size of "size" param and with 
# "hue" other color will appear for third param
sns.relplot(x="total_bill", y="tip", data=tips, size="size", hue="smoker", sizes=(50,200))
```

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
```

```python
df.tail()
```

```python
sns.relplot(x="time",y="value",kind="line",data=df,sort=True)
```

```python
df = pd.DataFrame(randn(500,2).cumsum(axis=0), columns=["time","value"])
```

```python
df.head()
```

```python
sns.relplot(x="time",y="value",kind="line",data=df,sort=True)
```

```python
sns.relplot(x="time",y="value",kind="line",data=df,sort=False)
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri,sort=True)
```

```python
# "ci" shows confidence interval
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri,ci=False)
```

```python
# To represent the spread of the distribution at each timepoint by plotting the 
# standard deviation instead of a confidence interval:
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri,ci="sd")
```
